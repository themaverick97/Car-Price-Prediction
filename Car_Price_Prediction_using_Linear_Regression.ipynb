```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.get_dummies(df,columns=['Fuel_Type','Seller_Type','Transmission'],drop_first=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X=df.drop(['Selling_Price','Car_Name'],axis=1)
y=df['Selling_Price']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=22)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
# Calculate error metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
adj_r2 = 1 - (1-r2)*(len(y)-1)/(len(y)-X.shape[1]-1)
```

```python
print(f"\nMean Squared Error:{mse}")
print(f"\nR-Squared:{r2} ")
print(f"\n Adjusted R-saqured: {adj_r2}")
```

```python
importance=model.coef_
```

```python
importance
```

```python
feature_importance=pd.DataFrame({'Features':X_train.columns,'Importance':importance})
feature_importance=feature_importance.sort_values(by='Importance',ascending=False)
print(feature_importance)
```

```python
residuals = y_test - y_pred
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(y_pred, residuals)
plt.title('Residual Plot')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.subplot(1, 2, 2)
sns.histplot(residuals, kde=True)
plt.title('Residual Distribution')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import  Ridge, Lasso, ElasticNet
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
X_train_scaled=sc.fit_transform(X_train)
X_test_scaled=sc.transform(X_test)
```

```python
linear_reg = LinearRegression()
linear_reg.fit(X_train_scaled, y_train)
y_pred_linear = linear_reg.predict(X_test_scaled)
```

```python
# Ridge Regression
ridge_reg = Ridge(alpha=1.0)
ridge_reg.fit(X_train_scaled, y_train)
y_pred_ridge = ridge_reg.predict(X_test_scaled)
```

```python
# Lasso Regression
lasso_reg = Lasso(alpha=1.0)
lasso_reg.fit(X_train_scaled, y_train)
y_pred_lasso = lasso_reg.predict(X_test_scaled)
```

```python
# ElasticNet Regression
elasticnet_reg = ElasticNet(alpha=1.0, l1_ratio=0.5)
elasticnet_reg.fit(X_train_scaled, y_train)
y_pred_elasticnet = elasticnet_reg.predict(X_test_scaled)
```

```python
def evaluate_performance(y_true, y_pred, model_name):
    mse = mean_squared_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    print(f"{model_name} - Mean Squared Error: {mse}")
    print(f"{model_name} - R-squared: {r2}")
```

```python
evaluate_performance(y_test, y_pred_linear, "Linear Regression")
evaluate_performance(y_test, y_pred_ridge, "Ridge Regression")
evaluate_performance(y_test, y_pred_lasso, "Lasso Regression")
evaluate_performance(y_test, y_pred_elasticnet, "ElasticNet Regression")
```

```python
# Calculate Sum of Squared Error (SSE) for each model
def calculate_sse(y_true, y_pred):
    return np.sum((y_true - y_pred) ** 2)

sse_linear = calculate_sse(y_test, y_pred_linear)
sse_ridge = calculate_sse(y_test, y_pred_ridge)
sse_lasso = calculate_sse(y_test, y_pred_lasso)
sse_elasticnet = calculate_sse(y_test, y_pred_elasticnet)

# Print SSE for each model
print(f"Linear Regression SSE: {sse_linear}")
print(f"Ridge Regression SSE: {sse_ridge}")
print(f"Lasso Regression SSE: {sse_lasso}")
print(f"ElasticNet Regression SSE: {sse_elasticnet}")
```
